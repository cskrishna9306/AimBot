```python
# Retrieve game data from riot games api
import requests

API_KEY = ''

# Riot Games API endpoint for summoner information
url = f"https://na.api.riotgames.com/val/content/v1/contents?locale=en-US"

# Include your API key in the request headers
headers = {
    "X-Riot-Token": API_KEY
}

# Make the request
response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    data = response.json()
    print(data)
else:
    print(f"Error: {response.status_code}, {response.text}")
```
